```python
import pandas as pd

from ICONProcessor.ICONMeteogram import ICONMeteogram
```

## Example: HEFEX study area

The following scenarios will be used:
* 51m resolution:
    * plot meteogram for station (full time frame)
    * plot meteogram for station (selected time frame)

```python
# folder that contains meteogram files
meteogram_dir = '/Users/geoalxx/Python/_ICON_data/hef/meteogram/'
```

### Show meteogram for one specific grid cell over entire modelling period

* Temperatur (2m)
* Surface Pressure
* Wind Speed
* Wind Gust
* Cloud Cover
* Sun hours

A timeframe can be defined as optional parameter. If empty, the maximum available timeframe will be used.

```python
met = ICONMeteogram(meteogram_dir,'METEOGRAM_2025')
met.plot_meteogram(location_idx=1)
```

Plot meteogram for different station and selected time frame

```python
met = ICONMeteogram(meteogram_dir,'METEOGRAM_2025')
met.plot_meteogram(location_idx=2, dt_from=pd.to_datetime('2025-08-06'), dt_to=pd.to_datetime('2025-08-07'))
```
